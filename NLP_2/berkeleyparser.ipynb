## 文分割や空行消すなどの前処理

```
import re

with open('./nlp.txt') as file_in, \
     open('./nlp.sen.txt', 'w') as file_out:
    for line in file_in:
        if line != '\n':
            line = re.sub(r'''
                         (?<=[\.|;|:|\?|!]) # 肯定の後読みで . or ; or : or ? or !
                         \s                 # 空白(改行への置換対象)
                         (?=[A-Z])          # 肯定の先読みで英大文字
                       ''', '\n', line, flags = re.VERBOSE)
            print(line.rstrip(), file=file_out)
```

```
!head nlp.sen.txt
```

```
!wc -l nlp.sen.txt
```
