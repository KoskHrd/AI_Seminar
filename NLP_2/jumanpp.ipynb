# 形態素解析演習

## 単語分割例

### 形態素解析器JUMAN++(JUMAN)と構文解析器KNPのPythonバインディングであるpyknpをインストール

```
!pip install pyknp
```

### 例

```
!echo "下鴨神社の参道は暗かった。" | jumanpp > sample.juman
```

```
from pyknp import Juman
juman = Juman()
data = ""
with open("sample.juman") as file_in:
    for line in file_in:
        data += line
        if line.strip() == "EOS":
            result = juman.result(data)            
            print(" ".join(mrph.midasi for mrph in result.mrph_list())) 
            data = ""
```

# 宿題1
下記のテキストを単語分割しなさい。
注意：1361行付近の「僕は速かに細君を書斎へ呼んだよ」の次の文でエラーが出ます。原因は次の文内の英文に, 空白区切りが含まれるからでした。try文などで`result = juman.result(data)`を囲って該当部分のパースを飛ばすか, 英文等のスペースを削除するとパースできます。

```
!wget http://www.cl.ecei.tohoku.ac.jp/nlp100/data/neko.txt
```

```
!head neko.txt
```

```
!cat neko.txt | jumanpp > neko.juman.txt
```

```
from pyknp import Juman
juman = Juman()
data = ""
with open("./neko.juman.txt") as file_in:
    for line in file_in:
        data += line
        try:
            if line.strip() == "EOS":
                result = juman.result(data)
                print(" ".join(mrph.midasi for mrph in result.mrph_list())) 
                data = ""
        except ValueError:
            print("ValueError : invalid literal for int() with base 10: '\\'")
```

# 宿題2
上記のテキストを単語分割しつつ品詞も出力してください。

```
from pyknp import Juman
juman = Juman()
with open("./neko.txt") as file_in:
    lines = file_in.readlines()
    for line in lines:
        line = line.replace("\n", "")
        if line == "":
            continue
        try:
            result = juman.analysis(line)
            print(" ".join(mrph.midasi for mrph in result.mrph_list()))
            print(" ".join(mrph.hinsi for mrph in result.mrph_list()))
        except ValueError:
            print("ValueError : invalid literal for int() with base 10: '\\'")
```
